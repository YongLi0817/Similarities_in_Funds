```python
today = datetime.today()
last_three_year = datetime(today.year-3, today.month,today.day - 7)
```

```python
ex = pd.read_excel('fund_intended.xlsx')
ex = [240004] + ex['基本信息'].tolist()
def f(x):
    x = str(x)
    if len(x) < 6:
        res = ['0' for i in range(6-len(x))]
        return ''.join(res) + x
    return x
ex = [f(x) for x in ex]
ex
```

```python
sql = f''' 
select * from fund_return_weekly where code in ({str(ex)[1:-1]}) and date > '{last_three_year}'
order by date
'''
df = pd.read_sql(sql, engine)
df['date'] = pd.to_datetime(df['date'])
# df = df.set_index('date')

df
```

```python
df = df.pivot(index='date',values='ret',columns='code')
df
```

```python
df2 = df.reset_index()
df2['year'] = df2['date'].dt.year
df2['qua'] = df2['date'].dt.quarter
def f(x):
    res = pd.DataFrame()
    for column in x.columns:
        
        if column == '240004' or column == 'date': continue
        # print(x[column])
        corr = x[column].corr(x['240004'], method='spearman')
        data = {'code':[column],'corr':[corr]}
        data = pd.DataFrame(data)
        res = pd.concat([res,data],axis=0)
    # print(res)
    return res

df2 = df2.groupby(['year','qua']).apply(lambda x:f(x)).reset_index()
df2['yearqtr'] = df2.year.astype(str)+' Q' +df2.qua.astype(str)
df2 = df2.pivot(index='code',columns='yearqtr',values='corr').dropna()
df2.drop(columns='2023 Q3',inplace=True)
df2['mean'] = df2.mean(axis=1)
df2
```

```python
from fastdtw import fastdtw
res = pd.DataFrame(columns=['code','整体corr','DTW'])
for column in df.columns:
    if column == '240004': continue
    spearman_corr = df[column].corr(df['240004'], method='spearman')
    dtw,path =  fastdtw(df[column], df['240004'])

    cosine_similarity = np.dot(df[column], df['240004']) / (np.linalg.norm(df[column]) * np.linalg.norm(df['240004']))

    res = pd.concat([res,pd.DataFrame({'code':[column],'整体corr':[spearman_corr],'DTW':dtw,'cosine_similarity':cosine_similarity})],axis=0)
res
```

```python
result = res.merge(df2,on='code',how='outer')
result
```
